# While-loop

With the help of the code, the loop with the condition is written as follows:

```
while <condition>:
   <instructions>
```
As long as the condition is met, the instructions are repeated

### Example:
(It iterates 5 times, until i=5, then it exits the loop)

````python
i = 0
while i < 5:
    print(i)
    i += 1
````

We can use boolean variable as a flag variable â€” with a value other than "yes", it signals the loop to stop:

````python
hungry = input("R u hungry? (y/n)")

while hungry == "y":
    print("Eat 1 more pancake")
    hungry = input("R u hungry? (y/n)")

print("Finally, get some rest")
````

We can limit the number of iterations using several conditions:

Imagine we have only 5 pancakes...

````python
n_pancakes = 5
n_eaten = 0
hungry = input("R u hungry? (y/n)")

while n_eaten < n_pancakes and hungry == "y":
    print("Eat 1 more pancake")
    n_eaten += 1
    hungry = input("R u hungry? (y/n)")

print(f'You have eaten {n_eaten} pancakes')
````

### Warning

If you forget to change your flag variable or for some other reason the condition always returns True, the loop will be infinite. Then just stop the code and correct the mistake.

### If-else in a while-loop

We can also check for some condition in the loop body

````python
n_pancakes = 5
n_eaten = 0
hungry = input("R u hungry? (y/n)")

while n_eaten < n_pancakes and hungry == "y":
    print("Eat 1 more pancake")
    n_eaten += 1

    if n_eaten == int(0.5 * n_pancakes):
        print("You have eaten half the pancakes. Stop.")

    hungry = input("R u hungry? (y/n)")

print(f'You have eaten {n_eaten} pancakes')
````

### While True - break variation

We can also use "break" to exit the loop

````python
n_pancakes = 5
n_eaten = 0
hungry = input("R u hungry? (y/n)")

while True:
    if n_eaten == n_pancakes or hungry != "y":
        break
    
    print("Eat 1 more pancake")
    n_eaten += 1

    if n_eaten == int(0.5 * n_pancakes):
        print("You have eaten half the pancakes. Stop.")

    hungry = input("R u hungry? (y/n)")

print(f'You have eaten {n_eaten} pancakes')
````

While-loop can be used to get multiple variables from keyboard
